## Hypothesis Testing: Difference in Total Crimes Between 2017 and 2022

Before delving into specific problem statements, we can perform a foundational hypothesis test to understand if there's a statistically significant change in overall crime levels across the years covered by our dataset. We will compare the mean total crimes in the earliest year (2017) with the latest year (2022).

**Null Hypothesis (H0):** There is no significant difference in the mean total number of crimes between the year 2017 and the year 2022.

**Alternative Hypothesis (H1):** There is a significant difference in the mean total number of crimes between the year 2017 and the year 2022.

```python
# First, ensure 'Total_Crimes' is calculated for the main 'df' DataFrame
# We'll use the column definitions that were set up earlier

identifier_cols_for_df_total = [
    'Row_ID', 'Year', 'State_Name', 'State_Code', 'District_Name',
    'District_Code', 'Police_Station_Circle'
]

# Get all columns in df that are not identifier columns, assuming they are crime counts
crime_cols_for_df_total = [col for col in df.columns if col not in identifier_cols_for_df_total]

df['Total_Crimes'] = df[crime_cols_for_df_total].sum(axis=1, skipna=True)

print("Calculated 'Total_Crimes' for the main DataFrame.")
display(df[['Year', 'Total_Crimes']].head())
```

```python
from scipy import stats

# Filter data for the years 2017 and 2022
crimes_2017 = df[df['Year'] == 2017]['Total_Crimes'].dropna()
crimes_2022 = df[df['Year'] == 2022]['Total_Crimes'].dropna()

# Perform an independent samples t-test
# We assume unequal variances (Welch's t-test) as a safer default
t_statistic, p_value = stats.ttest_ind(crimes_2017, crimes_2022, equal_var=False)

print(f"T-statistic: {t_statistic:.3f}")
print(f"P-value: {p_value:.3e}")

# Set significance level (alpha)
alpha = 0.05

if p_value < alpha:
    print(f"\nSince the p-value ({p_value:.3e}) is less than alpha ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference in the mean total number of crimes between 2017 and 2022.")
else:
    print(f"\nSince the p-value ({p_value:.3e}) is greater than alpha ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference in the mean total number of crimes between 2017 and 2022.")
```

## Interpretation of Results: Difference in Total Crimes Between 2017 and 2022
Test Performed: Independent Samples t-test (Welch's t-test)

Null Hypothesis (H0): There is no significant difference in the mean total number of crimes between the year 2017 and the year 2022.

## Hypothesis Testing: Urban vs. Rural Total Crimes

To address Problem Statement 3 more formally, we will conduct a hypothesis test to determine if there is a significant difference in the mean total number of crimes between urban and rural districts.

**Null Hypothesis (H0):** The mean total number of crimes in urban districts is equal to the mean total number of crimes in rural districts.

**Alternative Hypothesis (H1):** The mean total number of crimes in urban districts is not equal to the mean total number of crimes in rural districts.

```python
from scipy import stats

# Filter df_reduced for urban and rural districts
crimes_urban = df_reduced[df_reduced['Urban_Rural_Classification'] == 'Urban']['Total_Crimes'].dropna()
crimes_rural = df_reduced[df_reduced['Urban_Rural_Classification'] == 'Rural']['Total_Crimes'].dropna()

# Perform an independent samples t-test (Welch's t-test for unequal variances)
t_statistic_urban_rural, p_value_urban_rural = stats.ttest_ind(crimes_urban, crimes_rural, equal_var=False)

print(f"T-statistic (Urban vs. Rural Total Crimes): {t_statistic_urban_rural:.3f}")
print(f"P-value (Urban vs. Rural Total Crimes): {p_value_urban_rural:.3e}")

# Set significance level (alpha)
alpha = 0.05

if p_value_urban_rural < alpha:
    print(f"\nSince the p-value ({p_value_urban_rural:.3e}) is less than alpha ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference in the mean total number of crimes between urban and rural districts.")
else:
    print(f"\nSince the p-value ({p_value_urban_rural:.3e}) is greater than alpha ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference in the mean total number of crimes between urban and rural districts.")
```

### Interpretation of Results: Urban vs. Rural Total Crimes

This t-test compares the average total crimes in urban districts against rural districts.

*   **T-statistic**: Indicates the difference between the sample means relative to their variability.
*   **P-value**: The probability of observing such a difference if the null hypothesis (no difference) were true.

**Conclusion:** Based on the p-value, we will either reject or fail to reject the null hypothesis, providing statistical evidence for whether urban and rural areas have significantly different mean total crime levels.

## Hypothesis Test: Chi-squared Test for Association between Urban/Rural Status and 'Murder' Occurrence

To formally test for an association between the `Urban_Rural_Classification` and the occurrence of a specific crime type, we will use a Chi-squared test for independence. We'll use 'Murder' as an example crime type.

**Null Hypothesis (H0):** There is no association between a district's urban/rural status and the occurrence of 'Murder' crimes (i.e., they are independent).

**Alternative Hypothesis (H1):** There is an association between a district's urban/rural status and the occurrence of 'Murder' crimes (i.e., they are not independent).

```python
from scipy.stats import chi2_contingency

# Select the crime type to test
crime_type = 'Murder'

# Create a binary variable for crime occurrence (True if count > 0, False otherwise)
# We'll fill NA with 0 as NA in these columns means 0 crimes after initial replacement
df_reduced['Murder_Occurred'] = df_reduced[crime_type].fillna(0).astype(int) > 0

# Create a contingency table
contingency_table = pd.crosstab(df_reduced['Urban_Rural_Classification'], df_reduced['Murder_Occurred'])

print(f"Contingency Table for Urban/Rural vs. {crime_type} Occurrence:")
display(contingency_table)

# Perform the Chi-squared test
chi2, p_value_chi2, dof, expected = chi2_contingency(contingency_table)

print(f"\nChi-squared Statistic: {chi2:.3f}")
print(f"P-value: {p_value_chi2:.3e}")
print(f"Degrees of Freedom: {dof}")

# Set significance level (alpha)
alpha = 0.05

if p_value_chi2 < alpha:
    print(f"\nSince the p-value ({p_value_chi2:.3e}) is less than alpha ({alpha}), we reject the null hypothesis.")
    print(f"Conclusion: There is a statistically significant association between a district's urban/rural status and the occurrence of {crime_type} crimes.")
else:
    print(f"\nSince the p-value ({p_value_chi2:.3e}) is greater than alpha ({alpha}), we fail to reject the null hypothesis.")
    print(f"Conclusion: There is no statistically significant association between a district's urban/rural status and the occurrence of {crime_type} crimes.")
```
